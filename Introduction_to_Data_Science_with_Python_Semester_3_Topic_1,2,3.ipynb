Topic 3 Task 1 – Statistics of the insurance.csv dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np

# Load dataset
titanic = sns.load_dataset('titanic')  # or pd.read_csv('titanic.csv')

# Select only numerical columns
num_cols = titanic.select_dtypes(include=['float64', 'int64'])

# Define a function to compute all stats
def numerical_summary(df):
    summary = pd.DataFrame(index=df.columns)
    summary['Mean'] = df.mean()
    summary['Median'] = df.median()
    summary['Range'] = df.max() - df.min()
    summary['IQR'] = df.quantile(0.75) - df.quantile(0.25)
    summary['Std Dev'] = df.std()
    summary['Skewness'] = df.skew()
    summary['Kurtosis'] = df.kurtosis()
    return summary

# Compute summary
summary_stats = numerical_summary(num_cols)
print(summary_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical variables
num_cols = titanic.select_dtypes(include=['float64', 'int64']).columns

# Create subplots for each variable
for col in num_cols:
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    fig.suptitle(f'Distribution of {col}', fontsize=16)

    # Histogram
    sns.histplot(titanic[col].dropna(), kde=False, ax=axes[0], color='skyblue')
    axes[0].set_title('Histogram')

    # Density Plot
    sns.kdeplot(titanic[col].dropna(), ax=axes[1], fill=True, color='orange')
    axes[1].set_title('Density Plot')

    # Boxplot
    sns.boxplot(x=titanic[col], ax=axes[2], color='lightgreen')
    axes[2].set_title('Boxplot')

    plt.tight_layout()
    plt.show()
```

📌 age
Histogram/Density: Bell-shaped, slightly right-skewed.

Boxplot: Mild right skew, few outliers.

Stats: Skewness ≈ 0.4 → confirms slight right skew.

🧠 Description: Roughly normal, moderate spread, some outliers in upper range.

📌 fare
Histogram/Density: Highly right-skewed, long tail.

Boxplot: Extreme outliers (high fares).

Stats: Skewness > 4, Kurtosis > 30 → confirms strong right skew and heavy tails.

🧠 Description: Highly skewed, large variance, many outliers.

📌 sibsp and parch
Histogram: Discrete values (0,1,2,...), right-skewed.

Boxplot: Many zeros, long upper tail.

Stats: High skewness and kurtosis.

🧠 Description: Discrete, skewed, not normally distributed, family sizes are small for most.

📌 pclass
Histogram: Discrete 1/2/3, more people in 3rd class.

Boxplot: Limited but skewed distribution.

Stats: Skewness ≈ -0.6 → slight left skew.

🧠 Description: Categorical numeric, not truly continuous; visually skewed left due to 3rd class dominance.

📌 survived
Histogram: Binary values 0 or 1.

Boxplot: Not informative — binary.

Stats: Not meaningful for shape; purely a classification target.

🧠 Description: Binary target; visualization is only useful for proportions.

```python
from numpy.random import default_rng
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

seed = 1
rng = default_rng(seed=seed)

values = ["small", "medium", "large"]
N = 10000  # Try 100, 1000, then 10000
x = rng.normal(loc=100, scale=20, size=N)
y = rng.uniform(low=-5, high=5, size=N)
z = rng.choice(values, size=N, replace=True)

data = pd.DataFrame({'norm': x, 'unif': y, 'categ': z})

# Summary stats
print(data['norm'].mean(), data['norm'].std())

# Plot histograms
sns.histplot(data['norm'], kde=True)
plt.title(f"Normal distribution (N={N})")
plt.show()
```

N = 100: Histogram is bumpy and irregular.

N = 1000: Smoother, more bell-shaped.

N = 10000: Very smooth bell curve, closely resembling a normal distribution.

This is because more samples give better representation of the true distribution.

| N     | Mean (norm)  | Std Dev (norm) |
| ----- | ------------ | -------------- |
| 100   | ≈ 99–101     | ≈ 18–22        |
| 1000  | ≈ 99.5–100.5 | ≈ 19.5–20.5    |
| 10000 | ≈ 100        | ≈ 20           |

As N increases:

Mean → 100

Std → 20

This is due to the Law of Large Numbers, which says the sample statistics converge to population values as sample size increases.

| N     | Smoothness | Accuracy of Stats |
| ----- | ---------- | ----------------- |
| 100   | Rough      | Low               |
| 1000  | Medium     | Medium            |
| 10000 | Smooth     | High              |
